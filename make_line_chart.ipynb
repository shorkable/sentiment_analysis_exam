```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
df = pd.read_csv('data.csv')
df.head(3)

df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df[['Date', 'CP']].groupby(['Date'], as_index=False).mean()
df.head(3)
```

```python
target_date = pd.to_datetime('2021-05-06')

plt.plot(df['Date'], df['CP'])
plt.gcf().autofmt_xdate()
plt.grid(True, linestyle='--', linewidth=0.3, color='gray', alpha=0.3)
plt.axvline(x=target_date, color='red', linestyle=':', linewidth=1.5, label='Special Event')
plt.text(target_date, plt.ylim()[1] * 0.1, 'Train/Test split', color='red')

plt.title("S&P Closing Price over time")
```
